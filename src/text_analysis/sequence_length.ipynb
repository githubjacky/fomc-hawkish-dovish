# Load Data set

```python
from datasets import load_dataset

dataset = load_dataset("gtfintechlab/fomc_communication")
```

# Distribution of Sequences' Length

```python
from flair.data import Sentence
import matplotlib.pyplot as plt
import numpy as np
```

```python
def stats_sent_len(cate: str = "train"):
    sentence_lengths = []
    for i in range(len(dataset[cate])):
        sentence_lengths.append(len(Sentence(dataset[cate][i]["sentence"])))

    print("max: ", np.max(sentence_lengths))
    print("min: ", np.min(sentence_lengths))
    print("mean: ", np.mean(sentence_lengths))
    print("median: ", np.median(sentence_lengths))
    print("std: ", np.std(sentence_lengths))

    mean_value = np.mean(sentence_lengths)

    plt.hist(sentence_lengths, bins='auto', alpha=0.7, color='blue')
    plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label='Mean')
    plt.text(mean_value, max(plt.hist(sentence_lengths, bins='auto')[0]) * 0.97,
            f'Mean: {mean_value:.2f}', fontsize=20, color='red')
    plt.xlabel('Sequence Length')
    plt.ylabel('Count')
    plt.title('Distribution of Sequence Lengths (Train)')

    return plt
```

```python
plt1 = stats_sent_len("train")
```
